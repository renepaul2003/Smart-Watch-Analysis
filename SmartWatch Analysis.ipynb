#  Smart Watch Analaysis

Smartwatches have become ubiquitous, and they generate a vast
amount of data related to users’ health and fitness. In this project,
we will dive into the world of smartwatch data analysis using
Python. We’ll explore how to collect, clean, and analyze data from
smartwatches, gaining valuable insights into users’ activity patterns,
sleep quality, and more.

The dataset contains 940 entries and 15 columns, with the following variables:

* Id: Unique identifier for each user.
* ActivityDate: Date of the recorded actitivty.
* TotalSteps: Total Steps taken on that day
* TotalDistance: Total distance covered
* LoggedActivitiesDistance: Distance from manually logged activities.
* VeryActiveDistance, ModeratelyActiveDistance, LightActiveDistance,SedentaryActiveDistance: Distances covered at different              activity levels.
* VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes: Minutes spent in various activity levels.
* Calories: Calories burned.

### TRAINING  AND TESTING THE DATA

```python
# Spliting the data
from sklearn.model_selection import train_test_split
#assigning the dataset
x=data.drop(['Calories'],axis=1)
y=data['Calories']
# shapes of x and y
print(x.shape)
print(y.shape)
```

```python
# identifying the shape of xtrain,ytrain,xtest,ytest
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

###  Model Devolpment

There are so many state-of-the-art ML models available in academia but some model fits better to some problem while some fit better than other. So, to make this decision we split our data into training and validation data. Then we use the validation data to choose the model with the highest performance

In this regression problem ,we use these models

* LinearRegression
* SupportVectorRegressor(SVR)
* RandomForestRegressor
* GradientBoostingRegressor
* KNeighborsRegressor

To predict the accuracy we will use the accuracy score function from scikit-learn library.

```python
# importing the model LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
```

```python
# fitting xtrain ,ytrain into model Linear Regression
lr.fit(xtrain,ytrain)
```

```python
# predicting the training and testing dataset
ypredtr=lr.predict(xtrain)
ypredte=lr.predict(xtest)
#predicting  the score of model Linear Regression 
print(r2_score(ytrain,ypredtr))
print(r2_score(ytest,ypredte))
print(cross_val_score(lr,x,y,cv=5,).mean())
```

```python
#importing the model SVR
from sklearn.svm import SVR
svr=SVR()
# fitting xtrain,ytrain into model
svr.fit(xtrain,ytrain)
```

```python
# predicting the training and testing dataset
ypredtrain=svr.predict(xtrain)
ypredtest=svr.predict(xtest)
#predicting the score of the model SVR
print(r2_score(ytrain,ypredtrain))
print(r2_score(ytest,ypredtest))
print(cross_val_score(svr,x,y,cv=5,).mean())
```

This model is not good

```python
# importing the model RandomForest Regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
#fitting the xtrain,ytrain into model RandomForestRegressor
rf.fit(xtrain,ytrain)
```

```python
# predicting the training and testing dataset
ypredtrain1=rf.predict(xtrain)
ypredtest1=rf.predict(xtest)
# predicting the scxore of the model Random Forest Regressor
print(r2_score(ytrain,ypredtrain1))
print(r2_score(ytest,ypredtest1))
print(cross_val_score(rf,x,y,cv=5,).mean())
```

```python
# importing the model GradientBoosting
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor()
# fitting xtrain,ytrain into the model GradientBoostingRegressor
gb.fit(xtrain,ytrain)
```

```python
# predicting the training and testing dataset
ypredtrain3=gb.predict(xtrain)
ypredtest3=gb.predict(xtest)
# predicting the score of the model Gradientboosting regressor
print(r2_score(ytrain,ypredtrain3))
print(r2_score(ytest,ypredtest3))
print(cross_val_score(gb,x,y,cv=5,).mean())
```

```python
# importing the model KNeighorsRegressor
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
#fitting xtrain,ytrain into the model KNeighborsregressor
kn.fit(xtrain,ytrain)
```

```python
ypredtrain4=kn.predict(xtrain)
ypredtest4=kn.predict(xtest)
#predicting the score of the model KNeighborsregressor
print(r2_score(ytrain,ypredtrain4))
print(r2_score(ytest,ypredtest4))
print(cross_val_score(kn,x,y,cv=5,).mean())
```
